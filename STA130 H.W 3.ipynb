```python
import pandas as pd
import plotly.express as px

# Load the dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pd.read_csv(url)

# Create the histogram
fig = px.histogram(
    df, 
    x="flipper_length_mm", 
    color="species", 
    title="Flipper Length Distribution by Penguin Species",
    labels={"flipper_length_mm": "Flipper Length (mm)", "species": "Species"},
    nbins=30
)

# Calculate statistics for each species
species_stats = df.groupby('species')['flipper_length_mm'].agg(
    ['mean', 'median', 'std', 'min', 'max', 'quantile']
).rename(columns={'quantile': 'q25', 'quantile': 'q75'})

# Calculate additional quantiles
species_stats['q25'] = df.groupby('species')['flipper_length_mm'].quantile(0.25)
species_stats['q75'] = df.groupby('species')['flipper_length_mm'].quantile(0.75)

# Add vertical lines and rectangles for range, IQR, and 2σ from the mean
for species, stats in species_stats.iterrows():
    mean_value = stats['mean']
    median_value = stats['median']
    std_value = stats['std']
    min_value = stats['min']
    max_value = stats['max']
    q25_value = stats['q25']
    q75_value = stats['q75']
    std_lower = mean_value - 2 * std_value
    std_upper = mean_value + 2 * std_value

    # Add lines for min and max (Range)
    fig.add_vline(x=min_value, line=dict(color='green', width=1, dash='solid'))
    fig.add_vline(x=max_value, line=dict(color='green', width=1, dash='solid'))

    # Add rectangle for IQR (25th to 75th percentile)
    fig.add_vrect(
        x0=q25_value, x1=q75_value,
        fillcolor='orange', opacity=0.2, line_width=0,
        annotation_text=f"IQR ({species})", annotation_position="top left"
    )

    # Add rectangle for 2 standard deviations from mean
    fig.add_vrect(
        x0=std_lower, x1=std_upper,
        fillcolor='purple', opacity=0.2, line_width=0,
        annotation_text=f"2σ Range ({species})", annotation_position="bottom left"
    )

# Show the plot
fig.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pd.read_csv(url)

# Set the style of the visualization
sns.set(style="whitegrid")

# Create a figure
plt.figure(figsize=(12, 8))

# Generate a color palette
palette = sns.color_palette("husl", n_colors=df['species'].nunique())

# Create the histogram with KDE for flipper_length_mm by species
sns.histplot(
    data=df, 
    x="flipper_length_mm", 
    hue="species", 
    kde=True, 
    element="step", 
    stat="density", 
    common_norm=False, 
    palette=palette
)

# Calculate statistics for each species
for species in df['species'].unique():
    species_data = df[df['species'] == species]['flipper_length_mm']
    
    mean_value = species_data.mean()
    std_dev = species_data.std()
    min_value = species_data.min()
    max_value = species_data.max()
    q1 = species_data.quantile(0.25)
    q3 = species_data.quantile(0.75)
    iqr = q3 - q1
    
    # Mean and median lines
    plt.axvline(mean_value, color='red', linestyle='--', linewidth=2, label=f'{species} Mean' if species == df['species'].unique()[0] else "")
    plt.axvline(species_data.median(), color='blue', linestyle='--', linewidth=2, label=f'{species} Median' if species == df['species'].unique()[0] else "")
    
    # Range lines
    plt.axvline(min_value, color='green', linestyle=':', linewidth=2, label=f'{species} Min' if species == df['species'].unique()[0] else "")
    plt.axvline(max_value, color='orange', linestyle=':', linewidth=2, label=f'{species} Max' if species == df['species'].unique()[0] else "")
    
    # IQR lines
    plt.axvline(q1, color='purple', linestyle='-.', linewidth=2, label=f'{species} Q1' if species == df['species'].unique()[0] else "")
    plt.axvline(q3, color='brown', linestyle='-.', linewidth=2, label=f'{species} Q3' if species == df['species'].unique()[0] else "")
    
    # Standard deviation lines
    plt.axvline(mean_value - 2 * std_dev, color='cyan', linestyle=':', linewidth=2, label=f'{species} -2 Std Dev' if species == df['species'].unique()[0] else "")
    plt.axvline(mean_value + 2 * std_dev, color='magenta', linestyle=':', linewidth=2, label=f'{species} +2 Std Dev' if species == df['species'].unique()[0] else "")

# Add titles and labels
plt.title('Distribution of Flipper Length by Species with Statistical Indicators', fontsize=16)
plt.xlabel('Flipper Length (mm)', fontsize=14)
plt.ylabel('Density', fontsize=14)

# Create custom legend labels
handles = [plt.Line2D([0], [0], color=palette[i], lw=4) for i in range(len(df['species'].unique()))]
plt.legend(handles + 
           [plt.Line2D([0], [0], color='red', lw=2, linestyle='--', label='Mean'),
            plt.Line2D([0], [0], color='blue', lw=2, linestyle='--', label='Median'),
            plt.Line2D([0], [0], color='green', lw=2, linestyle=':', label='Min'),
            plt.Line2D([0], [0], color='orange', lw=2, linestyle=':', label='Max'),
            plt.Line2D([0], [0], color='purple', lw=2, linestyle='-.', label='Q1'),
            plt.Line2D([0], [0], color='brown', lw=2, linestyle='-.', label='Q3'),
            plt.Line2D([0], [0], color='cyan', lw=2, linestyle=':', label='-2 Std Dev'),
            plt.Line2D([0], [0], color='magenta', lw=2, linestyle=':', label='+2 Std Dev')],
           df['species'].unique().tolist() + ['Mean', 'Median', 'Min', 'Max', 'Q1', 'Q3', '-2 Std Dev', '+2 Std Dev'], 
           title='Species', 
           title_fontsize='13', 
           fontsize='12', 
           bbox_to_anchor=(1.05, 1), 
           loc='upper left')

# Show the plot with improved layout
plt.tight_layout()
plt.grid(True)
plt.show(renderer='png')
```

Q5: The primary difference between mean and median is that mean deals with the average of the dataset, while the median is the middle value of a data set when ordered from ascending or descending order. There are three particular cases where the relationship between mean and median come to light. The first case is when the mean and median are the same or extremely close. This relationship indicates that the data is symmetrically distributed, similiarly found in a normal distribution

The second case is when the mean is greater than the median. This relationship indicates that the data will be skewed right. This is typically when the majority of the data is to the left of the mean, with a few large data points bringing the mean up. For example, if a majority of people in a 30 person class got a 60%, but 1-2 people got a 100%, then the median will likely lie near 60%, however the few outliers will drag the mean up. 

The third case is when the median is larger than the mean. This relationship indicates that the data will be skewed left. This is when the the majority of the data is to the right of the mean, with a few lower data points dragging the mean down. Take the previous example, but imagine the majority of people got a 90% and a few people got a 40%. The median will lie near 90%, but the low marks will bring the mean down.
