How will this analysis be conducted?
First, if an individual has a NaN in any of the variables above, we will not consider them.
Observe that each of the dependent variables has the same options in the given question. Those options being:
Very Closely, Somewhat Closely and Not At All. 
We will create a simple system to determine how strictly an individual followed covid guidelines.
For each variable, if the participant chose "Very Closely" they'll be assigned a score of 3.
If the participant chose "Somewhat Closely" they'll be assigned a score of 2
If the participant chose "Not At All" they'll be assigned a score of 1.
Then the scores of all scores will be added up to get a final score, a high score represents strictly following the guidelines, and a low score represents loosely following guidelines.

Furthermore, we will also place a score on our independent variable:
Never = 1
Almost Never = 2
Rarely = 3
Sometimes = 4
Often = 5
Very Often = 6
Always = 7

We can now plot the values. For all the people who chose never, they're x value will be 1 and the y. value will be the total score of the Covid guidelines based on the system we earlier defined. Similarilr, for people who chose almost never will have an x value of 2 and the corresponding total score. This process will continue for x values from 1-7. After that, we can generate a fitted linear regression to determine the relationship between feeling's of sickness/weakness and strictly following Covid-19 guidelines.


Possible Results:
I suspect that the results will indicate that those who often (or more than often) feel sick/weak will be the ones who strictly follow Covid 19 guidelines. While one may expect that those who feel sick often stems from their lack of certain hygenic rules or a high exposure to germs while outside their household, it's possible this isn't the case. I'd make the hypothesis that those who feel sick more often will be the ones to strictly follow the guidelines, as their susceptibility to being sick provides more incentive to avoid possible Covid exposure. Moreover, those who rarely feel sick will have less incentive to avoid possible Covid exposure, as their immune system is good enough such that any basic amount of exposure will leave them with little to no symptoms.

How is this helpful?

Covid guidelines simulated far more rigorous hygenic guidelines in order to avoid the spread of germs, and so this data could represent how those who still follow rigorous hygenic guidelines get sick more often than those who don't, which one can consider a contradiction to the typical standard.

I would suspect this demonstrates that feelings of loneliness stem from a lower number of close friends. Having a lower number of close friends would logically mean that the opportunity to spend time with people is also lower.

```python
import pandas as pd
url1 = 'https://raw.githubusercontent.com/pointOfive/stat130chat130/main/CP/var_names.csv'
url2 = 'https://raw.githubusercontent.com/pointOfive/stat130chat130/main/CP/CSCS_data_anon.csv'
cols = pd.read_csv(url1)
data = pd.read_csv(url2,
                   na_values=["9999", "", " ", "Presented but no response", "NA"])
empty = (data.isna().sum()==data.shape[0])
data = data[empty.index[~empty]] # keep non empty columns only

data.shape
```

```python
# This variable indicates cases we recommend removing from an analysis due to fast completion times or possible fraudulent responses.
data.REMOVE_case.value_counts()
```

```python
# Let's just keep the recommended data
dataV2 = data[data.REMOVE_case=='No'].copy()
dataV2.shape
```

```python
# And select out participants who are part of the cohort data 
# (but may also be a part of the cross-sectional data)
dataV2_cohort = dataV2[dataV2.SURVEY_cohort_participant].copy()
dataV2_cohort.shape
```

```python
# And remove columns that have less than some number of missing values
# We could consider no missing values using 850, or...?
missingness_limit = 100 # this retains 166 of 1024 columns that aren't fully empty
dataV2_cohortV2 = dataV2[dataV2.SURVEY_cohort_participant].copy()
columns2keep = dataV2_cohortV2.isna().sum() < missingness_limit
columns2keep = columns2keep.index[columns2keep]
dataV2_cohortV3 = dataV2_cohortV2[columns2keep].copy()
dataV2_cohortV3.shape
```
